```python
import pandas as pd
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

# Select your features (columns)

```python
df.describe()
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
y = df["koi_disposition"].values.reshape(-1, 1)
X = df.iloc[:, 1:]
```

```python
print(X.shape, y.shape)
```

```python
from sklearn.preprocessing import LabelEncoder

# Step 1: Label-encode data set
label_encoder = LabelEncoder()
label_encoder.fit(y)
encoded_y = label_encoder.transform(y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, encoded_y, random_state=1)
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
y_train = y_train.reshape(-1, 1)
y_test = y_test.reshape(-1, 1)
```

```python
# Scale your data
from sklearn.preprocessing import StandardScaler
X_scaler = StandardScaler().fit(X_train)

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

# Train the Model



```python
from tensorflow.keras.models import Sequential

model = Sequential()
```

```python
from tensorflow.keras.layers import Dense
number_inputs = 40
number_hidden_nodes = 4
model.add(Dense(units=number_hidden_nodes,
                activation='sigmoid', input_dim=number_inputs))
```

```python
number_classes = 3
model.add(Dense(units=number_classes, activation='softmax'))
model.add(Dense(units=number_classes, activation='softplus'))
```

```python
model.summary()
```

```python
# Use categorical crossentropy for categorical data and mean squared error for regression
# Hint: your output layer in this example is using software for logistic regression (categorical)
# If your output layer activation was `linear` then you may want to use `mse` for loss
model.compile(optimizer='Adamax',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# Fit (train) the model
model.fit(
    X_train_scaled,
    y_train,
    epochs=1000,
    shuffle=True,
    verbose=2
)
```

```python
# Evaluate the model using the testing data
model_loss, model_accuracy = model.evaluate(
    X_test_scaled, y_test, verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Save the model
model.save('neural_network', save_format='tf')
```

```python
# Load the model
from tensorflow.keras.models import load_model
old_model = load_model('neural_network')
```

```python
# Evaluate the old model using the testing data
model_loss, model_accuracy = old_model.evaluate(
    X_test_scaled, y_test, verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
